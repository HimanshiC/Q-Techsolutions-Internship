```python
import pandas as pd
ogdf=pd.read_csv("Salesstore.csv")
df=ogdf.copy()
```

```python
print(df.isnull().sum())
```

```python
print(df.columns)
```

```python
df.columns=df.columns.str.strip()          
df.columns=df.columns.str.lower()
print(df.shape)
print(df.columns)
```

```python
print(df.dtypes)
```

```python
print(df.nunique())
```

```python
print("The descriptive statistics are")
print("\n",df.describe())
print("Mean of sales data",df['sales'].mean())
print("Mean of profit data",df['profit'].mean())
print("The customer with most orders",df['customer_name'].mode())
print("The region with most orders", df['region'].mode())
```

```python
std_sales = df['sales'].std()
std_profit = df['profit'].std()
min_sales = df['sales'].min()
max_sales = df['sales'].max()
range_sales = max_sales - min_sales
print("The standard deviation for \n sales=",std_sales,"\t profit=",std_profit)
print("The minimum sales is",min_sales,"\n The maximum sales is",max_sales,"\n The range of the sales is",range_sales)
```

```python
cols=['sales', 'profit', 'order_quantity']
for col in cols:
            print(col,": mean:", df[col].mean(),
            "median:", df[col].median(),
            "std:", df[col].std(),
            "skew:", df[col].skew(),
            "kurtosis:", df[col].kurtosis())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(18,6))

plt.subplot(1,3,1)
sns.boxplot(x=df['profit'], color='lightgreen')
plt.title('profit Outliers')

plt.subplot(1,3,2)
sns.boxplot(x=df['sales'], color='salmon')
plt.title('sales Outliers')

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='region', y='sales', data=df, estimator=sum, errorbar=None)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='product_category', y='profit', data=df, estimator=sum, errorbar=None)
plt.title('Total Profit by Category')
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
sns.histplot(df['sales'], kde=True, ax=axes[0])
axes[0].set_title('Sales Distribution')
axes[0].set_xlabel('Sales')

sns.histplot(df['profit'], kde=True, ax=axes[1])
axes[1].set_title('Profit Distribution')
axes[1].set_xlabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sales', y='profit', hue='region', data=df)
plt.title('Sales vs Profit by Region')
plt.show()
```

```python
SCATTER PLOT FOR COMPARITIVE ANALYSIS
```

```python

region_summary = df.groupby('region').agg({'sales': ['mean', 'sum', 'count'],'profit': ['mean', 'sum']}).round(2)
    
category_summary = df.groupby('product_category').agg({'sales': ['mean', 'sum', 'count'],'profit': ['mean', 'sum']}).round(2)
    
```

```python
print("The region wise summary:",region_summary)
print("The category wise summary:",category_summary)
```

```python
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='coolwarm', fmt=".2f", cbar=True)
plt.title('Correlation Matrix')
plt.show()
```

```python
sns.countplot(x='order_priority',hue='region',data=df,palette='viridis')
plt.title('Count of Orders by Priority')
plt.xlabel('Order Priority')
plt.ylabel('Count')
plt.show()
```

```python
segment_region_sales = df.groupby(['region', 'customer_segment'])['sales'].sum().unstack()
segment_region_sales.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis')
plt.title('Sales by Customer Segment and Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.legend(title='Customer Segment')
plt.show()
```
